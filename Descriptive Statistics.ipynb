## II. CÁC NHẬN ĐỊNH

1. HOME VALUE
- Giá nhà dựa trên khảo sát ở 51 bang của nước Mỹ trung bình khoảng 173000 USD:
+ Central Tendency: Dựa theo biểu đồ Histogram, ta thấy một nửa các bang có giá nhà trung bình tập trung ở phân khúc giá từ    125 - 250 nghìn USD:
+ Các bang ngoại lệ với giá nhà rất rẻ (<125000 USD) và giá nhà trung bình ở phân khúc cao cấp (>300000 USD).
+ Variability: Giá nhà trung bình ở Mỹ có một sự biến động lớn giữa các bang. Điều này chủ yếu là do một vài bang có giá trị bất động sản rất cao so với phần còn lại, tạo nên một khoảng cách giá lớn so với giá nhà trung bình của cả nước (Gấp 2,5 - 3 lần).
+ Distribution (Phân phối lệch trái) Giá nhà ở Mỹ tập trung chủ yếu ở các phân khúc giá rẻ và trung bình (<300000 USD).

```python
fig,axes=plt.subplots(1,2,figsize=(20,5))

fig1=sns.histplot(data=df,x="Home Value",ax=axes[0])
fig1.set(title='Home Value',xlabel='Home Value (USD)',ylabel='State')

fig2=sns.boxenplot(data=df,x='Home Value',ax=axes[1])
fig2.set(title='Home Value',xlabel='Home Value (USD)')
plt.show()
```

2. HOUSEHOLD INCOME
- Thu nhập hộ gia đình trung bình ở các bang của nước Mỹ rơi vào khoảng 50000 USD.
+ Central Tendency: Dựa vào biểu đồ Histogram, ta thấy thu nhập bình quân hộ gia đình tại một nửa các bang của Mỹ giao động từ 46000 - 56000 USD: 
+ Cá biệt có các bang thu nhập hộ gia đình ở mức thấp, chỉ khoảng từ 37000 - 43000 USD
+ Trong khi đó một vài bang có mức thu nhập trung bình cao ở khoảng 63000 - 70000 USD
+ Distribution: Thu nhập hộ gia đình tại các bang ở Mỹ có thể nói tập trung chủ yếu ở mức thu nhập trung bình và tốt. 


```python
fig,axes=plt.subplots(1,2,figsize=(20,5))

fig1=sns.histplot(data=df,x="HH Inc",ax=axes[0])
fig1.set(title='Household Income',xlabel='Household Income (USD)',ylabel='State')

fig2=sns.boxenplot(data=df,x='HH Inc',ax=axes[1])
fig2.set(title='Household Income',xlabel='Household Income (USD)')
plt.show()
```
